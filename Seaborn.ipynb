# Line Plot

```python
var1=[1,2,3,4,5,6,7]
var2=[4,3,2,5,1,7,6]
x=pd.DataFrame({"var1":var1,"var2":var2})
sns.lineplot(x="var1",y="var2",data=x)
plt.show()
```

```python
plt.plot(var1,var2)
```

```python
var1=[1,2,3,4,5,6,7]
var2=[4,3,2,5,1,7,6]
sns.lineplot(x=var1,y=var2) #without using dataset.
plt.show()
```

```python
a=sns.load_dataset("mpg").head(20)
a
```

```python
sns.lineplot(x="mpg",y="cylinders",data=a,hue="model_year",style="model_year",palette="ocean",legend=False)  #hue is used to make graph with respect  to/
plt.grid()
plt.title("Graph")
plt.show()
```

```python
sns.lineplot(a["cylinders"])
plt.show()
```

```python
sns.displot(a["cylinders"])
plt.show()
```

# Histogram

```python
var=sns.load_dataset("penguins")
var
```

```python
sns.displot(var["bill_length_mm"])
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],kde=True)
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240]) #to organise the data within limit we use bins
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True) #By using kde=True a line is emmeded over the bar graph.
plt.show()
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color="r") 
```
