```python
def fibonacci(n):
    fib_sequence = [0, 1]

    while len(fib_sequence) < n:
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])

    return fib_sequence[:n]

n = 10
result = fibonacci(n)
print(f"The first {n} Fibonacci numbers are: {result}")
```

```python
import cv2
import tkinter as tk
from PIL import Image, ImageTk

class CameraApp:
    def __init__(self, root, window_title):
        self.root = root
        self.root.title(window_title)

        # Create a label to display the video feed
        self.label = tk.Label(root)
        self.label.pack(padx=10, pady=10)

        # Create a button to start/stop the camera
        self.btn_start_stop = tk.Button(root, text="Start Camera", command=self.toggle_camera)
        self.btn_start_stop.pack(pady=10)

        # Initialize the camera
        self.cap = cv2.VideoCapture(0)  # 0 represents the default camera (you can change it based on your setup)
        self.video_feed()

    def toggle_camera(self):
        if self.btn_start_stop["text"] == "Start Camera":
            self.btn_start_stop["text"] = "Stop Camera"
            self.video_feed()
        else:
            self.btn_start_stop["text"] = "Start Camera"
            self.cap.release()

    def video_feed(self):
        ret, frame = self.cap.read()
        if ret:
            # Convert the OpenCV BGR image to RGB
            rgb_image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            # Convert the image to a PhotoImage object
            img = Image.fromarray(rgb_image)
            imgtk = ImageTk.PhotoImage(image=img)

            # Update the label with the new image
            self.label.imgtk = imgtk
            self.label.configure(image=imgtk)

            # Repeat the process after a delay (in milliseconds)
            self.root.after(10, self.video_feed)
        else:
            print("Error capturing frame")
            

if __name__ == "__main__":
    root = tk.Tk()
    app = CameraApp(root, "Camera Access App")
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

class FraudDetectionApp:
    def __init__(self, root, window_title):
        self.root = root
        self.root.title(window_title)

        # Create entry widgets for credit card information
        self.label_card_number = tk.Label(root, text="Card Number:")
        self.label_card_number.pack(pady=5)
        self.entry_card_number = tk.Entry(root)
        self.entry_card_number.pack(pady=5)

        self.label_expiration_date = tk.Label(root, text="Expiration Date (MM/YY):")
        self.label_expiration_date.pack(pady=5)
        self.entry_expiration_date = tk.Entry(root)
        self.entry_expiration_date.pack(pady=5)

        self.label_cvv = tk.Label(root, text="CVV:")
        self.label_cvv.pack(pady=5)
        self.entry_cvv = tk.Entry(root)
        self.entry_cvv.pack(pady=5)

        # Create a button to check for fraud
        self.btn_check_fraud = tk.Button(root, text="Check for Fraud", command=self.check_fraud)
        self.btn_check_fraud.pack(pady=10)

    def check_fraud(self):
        card_number = self.entry_card_number.get()
        expiration_date = self.entry_expiration_date.get()
        cvv = self.entry_cvv.get()

        # Basic fraud detection rules (for demonstration purposes)
        if len(card_number) != 16 or not card_number.isdigit():
            self.show_message("Invalid card number. Please enter a 16-digit numeric card number.")
        elif len(expiration_date) != 5 or not expiration_date[:2].isdigit() or not expiration_date[3:].isdigit():
            self.show_message("Invalid expiration date. Please enter a valid date in MM/YY format.")
        elif len(cvv) != 3 or not cvv.isdigit():
            self.show_message("Invalid CVV. Please enter a 3-digit numeric CVV.")
        else:
            self.show_message("No fraud detected. Transaction approved.")

    def show_message(self, message):
        # Display a pop-up message
        messagebox.showinfo("Fraud Detection Result", message)

if __name__ == "__main__":
    root = tk.Tk()
    app = FraudDetectionApp(root, "Credit Card Fraud Detection App")
    root.mainloop()
```
