# Series

```python
x=[1,2,3,4]
var=pd.Series(x)
print(var)
print(type(var))
# Getting data from index number
print(var[2])
```

```python
# To change the index of the series according to your wish It can e done as
a=[1,2,3,4]
var=pd.Series(a,index=['a','b','c','d'])
print(var)
```

```python
# To change the datatype of the series.
var=pd.Series(a,dtype="float",index=[1,2,3,4],name="Python")
print(var)
```

```python
dic={"name":"Ansh","Marks":100,"Section":"G1"}
var=pd.Series(dic,name="My dictionary")
print(var)
```

```python
print(pd.Series(12))
```

```python
a=pd.Series(12,index=[1,2,3,4,5,6,7],dtype="float")
print(a)
print(type(a))
```

```python
a=pd.Series(12,index=[1,2,3,4,5,6,7,8])
b=pd.Series(12,index=[1,2,3,4])
print(a+b)
# NaN is Not a Number.
```
