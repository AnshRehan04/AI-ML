# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3]
y=[4,5,6]
plt.bar(x,y)
plt.show()
```

```python
x=["Python","C","Js"]
y=[75,85,90]
plt.xlabel("language",fontsize=20)
plt.ylabel("Popularity")
plt.title("Programming")
plt.bar(x,y)
```

```python
x=["Python","C","Js"]
y=[75,85,90]
plt.xlabel("language",fontsize=20)
plt.ylabel("Popularity")
plt.title("Programming")
plt.bar(x,y,width=0.4) # to change the width
```

```python
x=["Python","C","Js"]
y=[75,85,90]
plt.xlabel("language",fontsize=20)
plt.ylabel("Popularity")
plt.title("Programming")
c=["y","r","b"]
plt.bar(x,y,width=0.3,color=c) # to change the color
```

```python
x=["Python","C","Js"]
y=[75,85,90]
plt.xlabel("language",fontsize=20)
plt.ylabel("Popularity")
plt.title("Programming")
c=["y","r","b"]
# plt.bar(x,y,width=0.3,color=c,align="center") # to align the text at center of graph
plt.bar(x,y,width=0.3,color=c,align="edge")  # to align the text at the edge of the graph.
```

```python
x=["Python","C","Js"]
y=[75,85,90]
plt.xlabel("language",fontsize=20)
plt.ylabel("Popularity")
plt.title("Programming")
c=["y","r","b"]
# plt.bar(x,y,width=0.3,color=c,align="center") # to align the text at center of graph
plt.bar(x,y,width=0.3,color=c,align="edge",edgecolor="black",linewidth=5,linestyle=":")  # to change the color of edge of the graph
# Linewidth is used to change the width of the edge border.
# Linestyle is used to change the style of the border of the edge of the graph.
```

```python
x=["Python","C","Js"]
y=[75,85,90]
plt.xlabel("language",fontsize=20)
plt.ylabel("Popularity")
plt.title("Programming")
c=["y","r","b"]
# plt.bar(x,y,width=0.3,color=c,align="center") # to align the text at center of graph
plt.bar(x,y,width=0.3,color=c,align="edge",edgecolor="black",linewidth=5,linestyle=":",alpha=0.4)   #alpha is used to blue the gaph slightly.
```

```python
x=["Python","C","Js"]
y=[75,85,90]
plt.xlabel("language",fontsize=20)
plt.ylabel("Popularity")
plt.title("Programming")
c=["y","r","b"]
# plt.bar(x,y,width=0.3,color=c,align="center") # to align the text at center of graph
plt.bar(x,y,width=0.3,color=c,align="center",edgecolor="black",label="Ansh")   #label is used to give label to the graph
plt.legend()
```

```python
# To make or compare two different bar graphs
x=["Python","C","Js"]
y=[75,85,90]
z=[20,30,10]
plt.xlabel("language",fontsize=20)
plt.ylabel("Popularity")
plt.title("Programming")
# c=["y","r","b"]
# plt.bar(x,y,width=0.3,color=c,align="center") # to align the text at center of graph
plt.bar(x,y,width=0.3,color="red",align="center",edgecolor="black",label="Ansh")   #label is used to give label to the graph
plt.bar(x,z,width=0.3,color="y",align="center",edgecolor="black",label="Ansh1")  
plt.legend()
```

# Scatter Plot

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,6,8,0,1,5,7,9]
plt.scatter(x,y,color="red")  #To change the color of dots
plt.title("Scatter Plot",fontsize=10) #To give the title of the graph
plt.xlabel("Day") #give label for x axis
plt.ylabel("Number") #give label for y axis
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,6,8,0,1,5,7,9]
size=[100,200,340,500,600,822,435,222,124]
plt.scatter(x,y,color="red",s=size,alpha=0.5)  #To change the size of dots
plt.title("Scatter Plot",fontsize=10) #To give the title of the graph
plt.xlabel("Day") #give label for x axis
plt.ylabel("Number") #give label for y axis
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,6,8,0,1,5,7,9]
size=[100,200,340,500,600,822,435,222,124]
plt.scatter(x,y,color="red",s=size,alpha=0.5,marker="*",edgecolor="black")  #To change the dots into star
plt.title("Scatter Plot",fontsize=10) #To give the title of the graph
plt.xlabel("Day") #give label for x axis
plt.ylabel("Number") #give label for y axis
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,6,8,0,1,5,7,9]
size=[100,200,340,500,600,822,435,222,124]
colors=["r","y","b","r","y","b","r","y","b"]
plt.scatter(x,y,c=colors,s=size,alpha=0.5,marker="*",edgecolor="black",linewidth=3)  #To change the width of edgeborder of stars.
plt.title("Scatter Plot",fontsize=10) #To give the title of the graph
plt.xlabel("Day") #give label for x axis
plt.ylabel("Number") #give label for y axis
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,6,8,0,1,5,7,9]
z=[12,45,3,22,11,33,44,5,6]
size=[100,200,340,500,600,822,435,222,124]
colors=["r","y","b","r","y","b","r","y","b"]
plt.scatter(x,y,c=colors,s=size,alpha=0.5)  #cmap is predefined property to give the colors.
plt.scatter(x,z,c=colors,s=size,alpha=0.5)
plt.colorbar()  #colorbar will 
plt.title("Scatter Plot",fontsize=10) #To give the title of the graph
plt.xlabel("Day") #give label for x axis
plt.ylabel("Number") #give label for y axis
plt.show()
```

# Pie Chart

```python
lang=["Python","C","C++"]
ratio=[80,30,40]
ex=[0.0,0.1,0.0]
plt.pie(ratio,labels=lang,explode=ex)
```

```python
lang=["Python","C","C++"]
ratio=[80,30,40]
ex=[0.0,0.1,0.0]
c=["r","b","y"]
plt.pie(ratio,labels=lang,explode=ex,colors=c,autopct="%0.2f%%")  #autopct is used to show the percentgae of given languages.
```

```python
lang=["Python","C","C++"]
ratio=[80,30,40]
ex=[0.0,0.1,0.0]
c=["r","b","y"]
plt.pie(ratio,labels=lang,explode=ex,colors=c,autopct="%0.2f%%",shadow=True,radius=0.5)  #shadow is used to give the shadow to the graphs..
# Radius is used to give radius to graphs
```

```python
lang=["Python","C","C++"]
ratio=[80,30,40]
ex=[0.0,0.1,0.0]
c=["r","b","y"]
plt.pie(ratio,labels=lang,explode=ex,colors=c,labeldistance=1.3)  #labeldistance is used to give the distance to the labels of the graphs
```

```python
lang=["Python","C","C++"]
ratio=[80,30,40]
ex=[0.0,0.1,0.0]
c=["r","b","y"]
plt.pie(ratio,labels=lang,explode=ex,colors=c,autopct="%0.2f%%",startangle=90,textprops={"fontsize":15})  #startangle is used to rotate the circle of the graph.
```

```python
lang=["Python","C","C++"]
ratio=[80,30,40]
ex=[0.0,0.1,0.0]
c=["r","b","y"]
plt.pie(ratio,labels=lang,explode=ex,colors=c,autopct="%0.2f%%")  #autopct is used to show the percentgae of given languages.
plt.title("Programming")
```
